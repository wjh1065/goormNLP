2. $\begin{bmatrix} 1 & 1 \\ 0 & 1 \\ 0 & 0 \end{bmatrix} \begin{bmatrix}x1\\x2\end{bmatrix} = \begin{bmatrix} 3 \\ 1 \\ 0\end{bmatrix}$는 해가 존재한다. \\

```python
import sympy as sy
from sympy.solvers.solveset import linsolve 

x1, x2 = sy.symbols('x1 x2')
A = sy.Matrix([[1,1],[0,1],[0,0]])
b = sy.Matrix([3,1,0])
system = A,b
linsolve(system, x1, x2)
```

7. A,B가 각각 invertible하면, $(A+B)^{-1} = A^{-1} + B^{-1}$ 이다.

```python
import numpy as np

A = np.array([[-2,1],[1,2]])
print(np.linalg.det(A))
A_inv = np.linalg.inv(A)

B = np.array([[1,-2],[1,2]])
print(np.linalg.det(B))
B_inv = np.linalg.inv(B)

# 1
print(np.linalg.inv(A + B))
print(A_inv + B_inv)
if np.array_equal(np.linalg.inv(A + B),A_inv + B_inv):
    print('same')
else:
    print('not same')
```

A: False / (A+B)의 inv와 각 A와 B의 inv를 더하고 두 행렬을 비교한 결과 서로 다르다.
